Que 1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
- Anomaly detection is a data mining process used to identify rare items, events, or observations that deviate significantly from a dataset's normal behavior. It acts as an early warning system to flag potential issues like technical glitches, fraudulent transactions, or security breaches. The three main types of anomalies are point, contextual, and collective.
- **Types of Anomalies with Examples**
    - **Point Anomalies:** A single data point that lies far outside the distribution of the rest of the data. It is the simplest type of anomaly.
        - **Example:** A credit card transaction for $10,000 when your typical transactions are under $100.
        - **Example:** A sudden spike in temperature reading (e.g., (100째C)) amidst a consistent data stream (e.g., (50째C)).
    - **Contextual Anomalies:** A data point that is considered anomalous only within a specific context, but might be considered normal in a different context. These are common in time-series data.
        - **Example:** Spending $500 on a Sunday might be normal, but spending $500 at 3:00 AM on a Tuesday might be abnormal.
        - **Example:** A temperature of 85째F(29째C) is normal in summer but considered an anomaly during winter.
    - **Collective Anomalies:** A set of data points that are considered anomalous when analyzed together, even if each individual point in the set does not seem unusual on its own.
        - **Example:** A sequence of low-volume, failed login attempts from different IP addresses over an hour, which together indicate a distributed, coordinated attack.
        - **Example:** A sudden increase in network traffic that only lasts for a few seconds might not be flagged alone, but if it happens repeatedly in a specific pattern, it may indicate a data exfiltration attempt.

Que 2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
- Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are all unsupervised machine learning algorithms used for anomaly detection, but they operate based on entirely different principles: isolating, clustering, and local density comparison, respectively.
- **1. Isolation Forest (iForest) :** Instead of profiling "normal" data, Isolation Forest isolates anomalies using binary trees. Anomalies are rare and distinct, requiring fewer splits to isolate (shorter path lengths) compared to normal points.
    - **Key Characteristics:**
        - **Algorithm Type:** Tree-based ensemble.
        - **Scalability:** Highly efficient with O(n log n) time complexity, making it suitable for massive, high-dimensional datasets.
        - **Robustness:** Insensitive to irrelevant features and scaling.
    - **Suitable Use Cases:**
        - **High-Volume Data:** Large-scale fraud detection in finance.
        - **High-Dimensional Data:** Cybersecurity network traffic analysis.
        - **Real-time Monitoring:** Industrial IoT sensor data.
- **2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise):** DBSCAN is a density-based clustering algorithm that groups points in dense regions and marks points in low-density areas as noise (outliers). It defines clusters by connecting points within a specified distance (epsilon) and density threshold (minPoints).
    - **Key Characteristics:**
        - **Algorithm Type:** Density-based clustering.
        - **Shape Handling:** Capable of finding arbitrarily shaped, non-spherical clusters.
        - **Parameter Sensitive:** Requires careful tuning of eps and minPts.
    - **Suitable Use Cases:**
        - **Spatial Data:** Geographic mapping and geospatial clustering.
        - **Non-linear Structures:** Datasets with complex, irregular cluster shapes.
        - **Noisy Datasets:** Where data naturally contains significant noise.
- **3. Local Outlier Factor (LOF) :** LOF measures the "local density deviation" of a point relative to its neighbors. It identifies outliers by comparing the density of a point to the density of its \(k\)-nearest neighbors.
    - **Key Characteristics:**
        - **Algorithm Type:** Density-based, neighbor-based approach.
        - **Locality:** Excellent at finding "local" anomalies that are not obvious globally.
        - **Computational Cost:** High complexity \(O(n^{2})\), making it less suitable for very large datasets.
    - **Suitable Use Cases:**
        - **Subtle/Local Anomalies:** Detecting outliers that exist within a small cluster of data, which might not be outliers globally.
        - **Varying Density Data:** Datasets where different clusters have significantly different densities.
        - **Data Cleaning:** Identifying subtle anomalies in small-to-medium sized, specialized datasets.

Que 3. What are the key components of a Time Series? Explain each with one example.
- Time series data consists of observations recorded over time, broken down into four key components used to analyze patterns and forecast future behavior: Trend (long-term direction), Seasonality (repeating short-term cycles), Cyclical (long-term oscillations), and Irregular (random, unpredictable fluctuations).
- **Trend (T):** Represents the long-term, consistent upward or downward movement in the data over a long period.
    - **Example:** A steady, year-over-year increase in global internet usage over the past decade.
- **Seasonality (S):** Regular, predictable, and repeating fluctuations that occur within a fixed, short period (e.g., daily, weekly, or annually).
    - **Example:** Increased ice cream sales during summer months, specifically every June-August.
- **Cyclical (C):** Oscillations in the data that repeat over a longer, non-fixed period (usually more than a year), often linked to economic cycles.
    - **Example:** The multi-year business boom and recession cycle.
- **Irregular (I or Noise):** Unpredictable, random, or residual variations caused by unexpected, rare, or unforeseen events.
    - **Example:** A sudden, sharp drop in stock prices due to a natural disaster or unexpected political news.

Que 4. Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
- A stationary time series has statistical properties - mean, variance, and autocorrelation that remain constant over time, ensuring the underlying process does not change. Non-stationary data, often having trends or seasonality, can be made stationary by transforming it using differencing, log transformations, or de-trending.
1. **Definition of Stationarity :** A stationary process is one where the mean and variance do not change over time, and the covariance between two periods depends only on the distance between them, not the absolute time.
    - **Strict Stationarity:** Entire distribution is constant over time.
    - **Weak Stationarity (Wide-Sense):** Mean and covariance are constant.
2. **Testing for Stationarity :**
    - **Visual Inspection:** Plotting the data to check for upward/downward trends or changing variance.
    - **Augmented Dickey-Fuller (ADF) Test:** A unit root test where the Null Hypothesis (\(H_{0}\)) assumes non-stationarity. A p-value \(\le 0.05\) indicates the series is likely stationary.
    - **KPSS Test:** Tests for trend stationarity; \(H_{0}\) is that the series is stationary.
    - **Summary Statistics:** Comparing mean and variance across different time periods.
3. **Transforming Non-Stationary to Stationary :**
    - **Differencing:** Subtracting the previous observation from the current one (\(y_{t}^{\prime }=y_{t}-y_{t-1}\)) removes trends.
    - **Seasonal Differencing:** Differencing with a lag equal to the period (e.g., \(y_{t}-y_{t-12}\) for monthly data) to remove seasonality.
    - **Logarithmic/Square Root Transformation:** Stabilizes non-constant variance (heteroscedasticity).
    - **De-trending:** Fitting and removing a linear or polynomial trend from the data.
    - **Rolling Statistics:** Using a moving average to reduce noise and identify the trend.

Que 5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
- These time series models differ primarily in their ability to handle trend, seasonality, and external variables. AR and MA are foundational, while ARIMA adds differencing for trends, SARIMA adds seasonal components, and SARIMAX includes external (exogenous) predictors, ranging from simple stationary modeling to complex, seasonal, and covariate-dependent forecasting.
- **AR (Autoregressive) Model:**
    - **Structure:** Uses linear combinations of past values (\(p\) lags) to predict future values.
    - **Application:** Stationary data where current values depend on previous values (e.g., short-term stock price trends).
- **MA (Moving Average) Model:**
    - **Structure:** Uses past forecast errors (\(q\) lags) to predict future values.
    - **Application:** Stationary data where shocks (unexpected events) affect future values (e.g., impact of a sudden news event on sales).
- **ARIMA (AutoRegressive Integrated Moving Average) Model:**
    - **Structure:** Combines AR, MA, and Differencing (\(d\) parameter) to achieve stationarity, denoted as \(ARIMA(p,d,q)\).
    - **Application:** Non-stationary, non-seasonal data with a trend (e.g., predicting total monthly electricity consumption).
- **SARIMA (Seasonal ARIMA) Model:**
    - **Structure:** Extends ARIMA by adding seasonal components (\(P,D,Q,s\)) to handle cyclical patterns, denoted as \(SARIMA(p,d,q)(P,D,Q)s\).
    - **Application:** Data with seasonality (e.g., monthly sales figures with yearly holiday spikes).
- **SARIMAX (Seasonal ARIMA with eXogenous variables) Model:**
    - **Structure:** Extends SARIMA by including exogenous regressors (\(X\)), which are external variables influencing the series.
    - **Application:** Situations where external factors, like promotions, temperature, or policy changes, impact the forecast (e.g., modeling sales based on price and advertising spend).

```python
# Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
# and decompose it into trend, seasonality, and residual components.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv('/content/AirPassengers.csv')

df['Month'] = pd.to_datetime(df['Month'], format='%Y-%m')
df.set_index('Month', inplace=True)

df.rename(columns={'#Passengers': 'Passengers'}, inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Passengers'], label='Original Series')
plt.title('Air Passengers Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()

decomposition = seasonal_decompose(df['Passengers'], model='additive', period=12)

plt.figure(figsize=(12, 8))
decomposition.plot()
plt.suptitle('Time Series Decomposition of Air Passengers', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```

```python
# Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies.
# Visualize the anomalies on a 2D scatter plot.

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import numpy as np
import seaborn as sns

df_taxi = pd.read_csv('/content/NYC_taxi_fare_data.csv')

features = ['fare_amount', 'trip_distance']
df_filtered = df_taxi[features].copy()

df_filtered.dropna(inplace=True)

df_filtered = df_filtered[(df_filtered['fare_amount'] > 0) & (df_filtered['fare_amount'] < 500)]
df_filtered = df_filtered[(df_filtered['trip_distance'] > 0) & (df_filtered['trip_distance'] < 100)]

print(f"DataFrame size after filtering: {len(df_filtered)} records")
display(df_filtered.describe())

model = IsolationForest(n_estimators=100, max_samples='auto', contamination=0.01, random_state=42)
model.fit(df_filtered[features])

df_filtered['anomaly'] = model.predict(df_filtered[features])

anomalies = df_filtered[df_filtered['anomaly'] == -1]
normal_data = df_filtered[df_filtered['anomaly'] == 1]

print(f"Number of anomalies detected: {len(anomalies)}")
print(f"Number of normal data points: {len(normal_data)}")

plt.figure(figsize=(12, 8))
sns.scatterplot(x='trip_distance', y='fare_amount', data=normal_data, color='blue', label='Normal Data', alpha=0.6)
sns.scatterplot(x='trip_distance', y='fare_amount', data=anomalies, color='red', label='Anomalies', s=100, marker='X')

plt.title('Anomaly Detection using Isolation Forest on NYC Taxi Fare Data')
plt.xlabel('Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 8: Train a SARIMA model on the monthly airline passengers dataset.
# Forecast the next 12 months and visualize the results.

import warnings
warnings.filterwarnings("ignore")
import itertools
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/AirPassengers.csv')

df['Month'] = pd.to_datetime(df['Month'], format='%Y-%m')
df.set_index('Month', inplace=True)
df.rename(columns={'#Passengers': 'Passengers'}, inplace=True)

order = (1, 1, 1)
seasonal_order = (1, 1, 1, 12)

print(f"Attempting to fit SARIMA model with order={order} and seasonal_order={seasonal_order}")

model = sm.tsa.statespace.SARIMAX(df['Passengers'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
results = model.fit()
print(results.summary())

pred_start_date = df.index[order[0] + seasonal_order[0]*seasonal_order[3] + order[1] + seasonal_order[1]*seasonal_order[3]]
predictions = results.predict(start=pred_start_date, end=df.index[-1], dynamic=False)

forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_ci = forecast.conf_int()

last_date = df.index[-1]
forecast_dates = pd.date_range(start=last_date, periods=forecast_steps + 1, freq='MS')[1:] # +1 and [1:] to exclude last_date
forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_dates)

plt.figure(figsize=(15, 7))
plt.plot(df.index, df['Passengers'], label='Original Series', color='blue')
plt.plot(predictions.index, predictions, label='SARIMA Fitted', color='green', linestyle='--')
plt.plot(forecast_series.index, forecast_series, label='SARIMA Forecast', color='red')

plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.3, label='95% Confidence Interval')

plt.title('Air Passengers: Original, Fitted, and SARIMA Forecast (Next 12 Months)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
# anomalies and visualize them using matplotlib.

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
import seaborn as sns
from sklearn.datasets import load_iris

iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

features_for_lof = ['sepal length (cm)', 'sepal width (cm)']
X = df_iris[features_for_lof]

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)

df_iris['lof_score'] = lof.fit_predict(X)
df_iris['lof_values'] = lof.negative_outlier_factor_

anomalies_lof = df_iris[df_iris['lof_score'] == -1]
normal_data_lof = df_iris[df_iris['lof_score'] == 1]

print(f"Number of anomalies detected by LOF: {len(anomalies_lof)}")
print(f"Number of normal data points by LOF: {len(normal_data_lof)}")

plt.figure(figsize=(10, 7))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=normal_data_lof, color='blue', label='Normal Data', alpha=0.6)
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=anomalies_lof, color='red', label='Anomalies', s=100, marker='X')

plt.title('Anomaly Detection using Local Outlier Factor (LOF) on Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

 How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?  
 Which time series model would you use for short-term forecasting (ARIMA /

 How would you validate and monitor the performance over time?  
 How would this solution help business decisions or operations?

**Answer**
- As a data scientist specializing in power grid monitoring, implementing a real-time (15-minute interval) energy demand forecasting and anomaly detection system requires a robust, automated pipeline. The workflow centers on using historical data, weather, and region-specific features to predict demand and identify anomalies, which typically indicate malfunctions, cyberattacks, or extreme consumption behavior.

- Here is the proposed real-time data science workflow:
1. **Anomaly Detection in Streaming Data :** For real-time 15-minute streaming data, Isolation Forest is the preferred choice for detecting abnormal spikes or drops.
    - **Why Isolation Forest:** It is highly efficient and scalable, making it ideal for streaming data scenarios where new data points must be scored quickly without needing to update the entire model. It operates by isolating anomalies rather than profiling normal behavior, allowing it to work well with high-dimensional data (e.g., combining time, weather, and consumption).
    - **Implementation:** The algorithm will compute an anomaly score for each new 15-minute data point. A threshold (e.g., based on the contamination rate or standard deviation) will flag anomalies.
    - **Alternative:** LOF (Local Outlier Factor) could be used if anomalies are subtle, but it is less efficient for real-time applications compared to Isolation Forest.
2. **Short-Term Forecasting Model :** For forecasting short-term energy demand (e.g., next 15 minutes to 24 hours), SARIMAX is the most robust choice.
    - **Why SARIMAX:** Energy data has high seasonality (daily/weekly cycles) and is heavily influenced by exogenous variables like temperature and humidity. SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors) handles both the seasonality and external inputs directly.
    - **Alternative:** SARIMA is suitable if only seasonality matters, but SARIMAX provides higher accuracy by incorporating weather forecasts as features.
    - **Advanced Option:** If the data exhibits highly non-linear relationships, an LSTM (Long Short-Term Memory) network would be used for deeper, non-linear pattern recognition, particularly for 15-minute intervals.
3. **Validation and Monitoring of Performance :** To ensure the model remains accurate, the system will use an automated monitoring framework.
    - **Validation:** A "walk-forward" validation approach will be used, where the model is retrained periodically (e.g., daily) using a rolling window of recent data.
    - **Metrics:** Performance will be tracked using RMSE (Root Mean Square Error) for forecasting accuracy and F1-Score for anomaly detection (balancing Precision and Recall).
    - **Monitoring:** Live dashboards will monitor the "drift" between predicted and actual demand. If the error exceeds a threshold (e.g., 3 standard deviations), an alert is triggered to retrain the model.
4. **Impact on Business Decisions and Operations :** This solution provides actionable insights that directly improve grid reliability and economics:
    - **Proactive Grid Maintenance:** Detecting anomalies in real-time allows operators to identify faulty equipment or stolen electricity immediately, reducing downtime and operational costs.
    - **Dynamic Load Management:** Accurate short-term forecasts prevent overproduction (wasted energy/money) or underestimation (blackouts), saving millions of dollars annually in grid management.
    - **Enhanced Reliability:** The combination of forecasting and anomaly detection strengthens the grid against cyber-physical attacks by identifying data-injection attacks or abnormal consumption patterns.
